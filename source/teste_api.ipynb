Como buscar um preço histórico no Yahoo Finance

```python
# Importar bibliotecas
import yfinance as yf
import pandas as pd
```

```python
# Definir o ativo que quero pesquisar. As ações brasileiras terminam em .SA
ticker = "^GSPC"
```

```python
# Chamar a API pedidno o histórico. O yf organiza esses dados num DF. 
# auto_adjust faz com que a coluna Adj Close apareça (que é o preço que soma dividendos e desdobramentos)

dados_brutos = yf.download(ticker, period="max", auto_adjust=False)
```

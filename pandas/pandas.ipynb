### Lambda 


```python
#Lambda com cálculos
my_lambda = lambda x: (x*2) + 3
print(my_lambda(15))
```

```python
#Lambda para strings
stringlambda = lambda x: x.lower()
print(stringlambda('Oie, ComO VocÊ estÁ?'))
```

```python
outra_lambda = lambda x: x[0] + x[-1]
print(outra_lambda("Como vai"))
```

```python
#Função Lambda
# a estrutura geralmente é:
# lambda x : [OUTCOME IF TRUE] if [CONDITIONAL] else [OUTCOME IF FALSE]
```

```python
#se fosse uma função. Ex horas extras acima de 40
def minha_funcao(x):
    if x > 40:
        return 40 + ((x - 40) * 1.5)
    else:
        return x
    
#quantas horas devo receber se trabalhar acima de 40?
print(minha_funcao(44))
print(minha_funcao(36))
print(minha_funcao(52))
```

```python
#Se for pra fazer a mesma função como Lambda, fica:
horas_lambda = lambda x: 40 + ((x - 40) * 1.5) if x > 40 else x
print(horas_lambda(44))
```

```python
#aplicar Lambda em colunas
df['Avaliação de 0 a 10'] = df['User Rating'].apply(lambda x: x * 2)
df
```

```python
#lambda para selecionar apenas o sobrenome
df['Sobrenome'] = df.Author.apply(lambda x: x.split(' ')[-1])
df
```

```python
#pode usar uma função lambda para uma linha inteira usando eow.column_name ou row['column_name]
# axis=1 aplica a função pras linhas
df['Classificação de nota'] = df.apply(lambda row:
    'Bem avaliado'
    if row['User Rating']  > 4.6
    else row['User Rating'],
    axis=1
)
df
```

### Renomeando Colunas

```python
#Renomear nome das colunas através da lista de nomes das colunas
exemplo_nome_colunas = pd.DataFrame({
    'name': ['John', 'Pedro', 'Paulo'],
    'age': [23,29,21]
})
exemplo_nome_colunas.columns = ['Nomes', 'Idade']
exemplo_nome_colunas
```

```python
#Renomear os nomes das colunas com .rename. usa o inplace=True para mudar o nomeno próprio dataframe e não no existente
exemplo_nome_colunas.rename(columns={
    'Nomes': 'Primeiro Nome',
    'Idade': 'Anos'},
    inplace=True
)
exemplo_nome_colunas
```

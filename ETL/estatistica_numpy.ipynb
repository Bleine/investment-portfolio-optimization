```python
import pandas as pd
import numpy as np
```

```python
ativos = pd.read_csv('dados_limpos.csv')
```

```python
ativos_pivotados = ativos.pivot(index='Date', columns='Ticker', values='Retorno Diário')
ativos_pivotados
```

```python
ativos_matrix = ativos_pivotados.to_numpy()
ativos_matrix
```

```python
estatistica = pd.DataFrame({
    'Média Diária': np.nanmean(ativos_matrix,axis=0),
    'Variância Diária': np.nanvar(ativos_matrix,axis=0),
    'Desvio Padrão Diário': np.nanstd(ativos_matrix,axis=0)
}, index = ativos_pivotados.columns)
print(estatistica)
```

```python
coeficiente_de_correlacao = np.corrcoef(ativos_matrix)
coeficiente_de_correlacao
```

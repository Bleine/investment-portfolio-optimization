```python
#Transformar números em porcentagem
def display_as_percentage(valor):
    return str(round(valor * 100, 2)) + '%'
teste = display_as_percentage(0.0023)
print(teste)
```

```python
#Simple Rate of Return - É uma porcentagem que indica a rentabilidade de um investimento ou projeto em um determinado período
def simple_return(start_price, end_price, dividend=0):
    return (end_price + dividend - start_price) / start_price
teste = simple_return(100,109,1)
teste = display_as_percentage(teste)
print(teste)
```

```python
#continuously compounded return
from math import log
def log_return(start_price, end_price):
    return log(end_price) - log(start_price)
teste = log_return(200,250)
teste = display_as_percentage(teste)
print(teste)
```

```python
#transformar a taxa diária em taxa anual, onde 't' = 252, pois quero transformar de retorno diário para anual
def retorno_anual(log_rate, t):
    return log_rate * t
teste = retorno_anual(0.001,252)
teste = display_as_percentage(teste)
print(teste)
```

```python
#Anual rate of return
def convert_returns(log_returns, t):
    return sum(log_returns) / len(log_returns) * t

retornos_teste = [-0.29,2.42,0.35,0.7,-1.23]
teste = convert_returns(retornos_teste, 252)
print(teste)
```

```python
#Variancia
import numpy as np
dataset = [-0.29,2.42,0.35,0.7,-1.23]
variancia = np.var(dataset)
print(variancia)
```

```python
#desvio padrão
import numpy as np
dataset = [-0.29,2.42,0.35,0.7,-1.23]
desvio_padrao = np.std(dataset)
print(desvio_padrao)
```

```python
#matriz de correlação
dataset_1 = [-0.29,2.42,0.35,0.7,-1.23]
dataset_2 = [-0.15,3.42,0.45,0.4,-2.23]
dataset_3 = [0.29,-2.42,-0.35,-0.7,1.23]
matriz_de_correlacao = np.corrcoef([dataset_1, dataset_2, dataset_3])
print(matriz_de_correlacao)
```
